## Подгонка распределения

```python
X = X['x']
```

```python
dfit = distfit(todf=True)
model = dfit.fit_transform(X)
```

```python
dfit.plot_summary()
plt.show()
```

```python
dfit.summary
```

В результате подгонки распределения - самое подходящее - это t-распределения с loc = 0.267, scale = 2.89. (Можно заметить, что это совпадает с mean и std, построенными по выборке)

```python
fig, ax = plt.subplots(1,2, figsize=(15, 6))
dfit.plot(chart='PDF', ax=ax[0])
dfit.plot(chart='CDF', ax=ax[1])
plt.show()
```

  Распределение данных: Данные хорошо описываются t-распределением, что указывает на близость к нормальному распределению, учитывая высокое значение степеней свободы.

  Доверительные интервалы: На графике показаны 95% доверительные интервалы (CII low и CII high), которые охватывают диапазон значений от примерно -5.0 до 10.0. Это указывает на то, что с вероятностью 95% истинное среднее значение находится в этом диапазоне.

  Плотность вероятности: Пик плотности вероятности находится около среднего значения (2.68), что подтверждает, что это наиболее вероятное значение в распределении.

  Частота значений: На гистограмме видно, что большинство значений сосредоточено вокруг среднего, с уменьшением частоты по мере удаления от него.

## Тест Колмогорова-Смирнова

```python
from scipy.stats import kstest

params = dfit.model['params']
dist_name = dfit.model['name']

D, p_value = kstest(X, dist_name, args=params)

print(f"Kolmogorov-Smirnov statistic (D): {D}")
print(f"p-value: {p_value}")

alpha = 0.05
if p_value > alpha:
    print("Гипотеза согласия не отвергается: данные соответствуют выбранному распределению.")
else:
    print("Гипотеза согласия отвергается: данные не соответствуют выбранному распределению.")
```
